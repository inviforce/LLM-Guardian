```python
import json
import os
import glob
import matplotlib.pyplot as plt

def plot_loss_from_checkpoints(save_dir, start_step=50):
    """
    Parses trainer_state.json from Hugging Face output folder and plots training loss,
    starting from a specific step to filter out early high-loss spikes.
    """
    # Look for trainer_state.json in root or subdirectories
    json_path = os.path.join(save_dir, "trainer_state.json")
    
    if not os.path.exists(json_path):
        checkpoint_files = glob.glob(os.path.join(save_dir, "checkpoint-*", "trainer_state.json"))
        if not checkpoint_files:
            print(f"Error: No 'trainer_state.json' found in {save_dir}")
            return
        # Get the latest checkpoint state file
        json_path = sorted(checkpoint_files, key=os.path.getmtime)[-1]

    print(f"[*] Reading log metrics from: {json_path}")
    
    with open(json_path, "r") as f:
        trainer_state = json.load(f)

    # Extract step and loss values starting from start_step
    steps = []
    losses = []
    
    for log in trainer_state.get("log_history", []):
        if "loss" in log and "step" in log:
            if log["step"] >= start_step:
                steps.append(log["step"])
                losses.append(log["loss"])

    if not steps:
        print(f"[-] No loss entries found for step >= {start_step}.")
        return

    # Plot graph
    plt.figure(figsize=(10, 5))
    plt.plot(steps, losses, marker='o', linestyle='-', color='#1f77b4', label='Train Loss')
    plt.title(f"SFT Training Loss Curve (From Step {start_step})")
    plt.xlabel("Step")
    plt.ylabel("Loss")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

# ======================================================
# Run Plotter
# ======================================================
SAVE_DIR = "../models/atbench_lora_final_v3_qwen"
plot_loss_from_checkpoints(SAVE_DIR, start_step=50)
```
